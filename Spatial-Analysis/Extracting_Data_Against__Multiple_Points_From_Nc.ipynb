```python
import xarray as xr
import pandas as pd
```

```python
# Read the NetCDF file
nc_file = "C:/Data/Wflow/DIK/DIK_Future/final_output_min_ppt_4.5/run_default/output.nc"
data = xr.open_dataset(nc_file)

# Read the CSV file
csv_file = "C:/Data/Wflow/DIK/CDM_drainage_points.csv"
coordinates = pd.read_csv(csv_file)
```

```python
data
```

```python
variable = data['q_river']
```

```python
variable.max()
```

```python
# Create an empty dataframe
df = pd.DataFrame()
df1 = pd.DataFrame()

# Extract data at each coordinate
for index, row in coordinates.iterrows():
    lat = row['Lat']
    lon = row['Long']


    point_data = variable.sel(lat=lat, lon=lon, method='nearest')

    # # Add the values to the dataframe
    column_name = f'({lat},{lon})'
    df = point_data.to_dataframe().reset_index()
    df1['date'] =df['time']
    df1[column_name] = df['q_river']
```

```python
df1.max()
```

```python
df1.to_csv("C:/Data/Wflow/DIK/DIK_Future/final_output_min_ppt_4.5/run_default/CDM_points.csv")
```
