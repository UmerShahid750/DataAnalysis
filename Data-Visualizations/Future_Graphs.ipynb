```python
# Filter DataFrame for May to July
jun_to_aug_df = df[df['month'].between(6, 8)]
jun_to_aug_df_obs = df1[df1['month'].between(6, 8)]
jun_to_aug_df
```

```python
yearly_max_future = jun_to_aug_df.groupby('year')[['SSP245_Mean','SSP245_Min','SSP245_Max','SSP585_Mean','SSP585_Min','SSP585_Max']].max().reset_index()
yearly_max_obs = jun_to_aug_df_obs.groupby('year')[['max_obs','min_obs','obs_mean']].max().reset_index()
yearly_max_future
```

```python
fig= plt.subplots(figsize=(20, 10))


x_title = 'year'
axis_font = 14

# baseline = df['Baseline']

# Plotting the lines for max and min values
plt.plot(yearly_max_future['year'], yearly_max_future['SSP245_Mean'], label='SSP245', color='#ff3232',alpha=1)
plt.plot(yearly_max_future['year'], yearly_max_future['SSP585_Mean'], label='SSP585', color='#660000',alpha= 1)

plt.plot(yearly_max_obs['year'], yearly_max_obs['obs_mean'], label='Baseline', color='black',lw=2,alpha=0.5)
# # plt.plot(x_values, max_values, label=f'SSP{case}_Max', color=colour,lw=3.5)

# # # Shading the area between max and min
plt.fill_between(yearly_max_future['year'], yearly_max_future['SSP245_Min'], yearly_max_future['SSP245_Max'], color='#ff3232', alpha=0.2)
plt.fill_between(yearly_max_future['year'], yearly_max_future['SSP585_Min'], yearly_max_future['SSP585_Max'], color='#660000', alpha=0.2)

plt.fill_between(yearly_max_obs['year'], yearly_max_obs['max_obs'], yearly_max_obs['min_obs'], color='grey', alpha=0.3)



# Adding labels, title, legend, etc.
plt.xlabel(x_title, fontsize = axis_font)
# Set x ticks at the specific x values of your data
plt.xticks(range(1990,2100,4))
plt.ylabel('Temperature [C]',fontsize= axis_font)
plt.title('Monsoon Maximum Tempeature ', fontsize=20)
plt.legend()
# Remove spines on the upper and right sides
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.grid(axis='y')
# Display the plot
plt.show()
```
