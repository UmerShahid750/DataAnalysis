```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import datetime as dt
import matplotlib.dates as mdates
import numpy as np
```

```python
path = r'C:\Users\LENOVO\OneDrive\WaterSprint\Projects\Badin-WASH Project\Climate Risk Assessment\validation'
path = path.replace("\\", "/")
path
```

```python
#Reading the corrected PMD files using pandas
Temp = pd.read_excel(path+'/Badin_Temperature.xlsx')
Temp['date'] = pd.to_datetime(Temp['date'])
Temp
```

```python
start_date = '2002-01-01'
end_date = '2012-12-31'
df1 = Temp[(Temp['date'] >= start_date) & (Temp['date'] <= end_date)]
df1
```

```python
plt.figure(figsize=(40, 15))
plt.subplots_adjust(bottom=0.15)
plt.xticks(rotation=45)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# Set the x-tick locations and labels
import matplotlib.dates as mdates
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=10)) # Show tick every month
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%Y')) # Set format of the tick
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=4)) # Show tick every month
# Create a plot with multiple lines
plt.plot('date', 'PMD_Max', data=df1, label='Observed', color='black')
plt.plot('date', 'ERA5_Max', data=df1, label='ERA5', color='red')

#Labelling the plot
plt.legend(loc='upper left')
plt.legend(fontsize='24')
plt.xlabel('Time',fontsize = '20')
plt.ylabel('Temperature in °C',fontsize = '25')
plt.title('Maximum Temperature Comparison', fontsize= '24')
```
